Scan the Iris dataset and give the inference as:
1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a boxplot for each feature in the dataset. 4. Compare distributions and identify outliers.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\OneDrive\Desktop\Desktop\New folder\Datasets\Iris.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Data Types:\n", df.dtypes)
print("\nUnique Species:", df['Species'].unique())
```

```python
sns.boxplot(df)
```

```python
q1=df["SepalWidthCm"].quantile(0.25)
q3=df["SepalWidthCm"].quantile(0.75)
iqr=q3-q1
lower=q1-iqr*1.5
upper=q3+iqr*1.5
df["SepalWidthCm"]=np.where(df["SepalWidthCm"]<lower,lower,df["SepalWidthCm"])
df["SepalWidthCm"]=np.where(df["SepalWidthCm"]>upper,upper,df["SepalWidthCm"])
plt.title(f"plt of SepalWidthCm")
sns.boxplot(data=df["SepalWidthCm"])
plt.show()
```

```python
features=df.drop(['Species'],axis=1)
features.head()
```

```python
for i in features:
    plt.title(f"Feature {i}")
    sns.boxplot(df,y=i,x="Species")
    plt.show()
```

```python
for i in features:
    plt.title(f"Feature {i}")
    sns.histplot(df,x=i,bins=10,hue="Species")
    plt.show()
```
