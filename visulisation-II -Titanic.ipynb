```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\OneDrive\Desktop\Desktop\New folder\Datasets\titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=["age","deck","embark_town"],inplace=True)
df.isnull().sum()
```

```python
sns.boxplot(df)
```

```python
lst=["pclass","sibsp","parch","fare"]
for i in lst:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    lower=q1-iqr*1.5
    upper=q3+iqr*1.5
    df[i]=np.where(df[i]<lower,lower,df[i])
    df[i]=np.where(df[i]>upper,upper,df[i])
    plt.title(f"plt of {i}")
    sns.boxplot(data=df[i])
    plt.show()
```
